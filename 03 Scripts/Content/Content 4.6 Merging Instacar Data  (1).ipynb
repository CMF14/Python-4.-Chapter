## Merging both Dataframe

```python
## Merging using argument order_id + indicator

df_merged_large = df_ords.merge(df_ords_prior, on = 'order_id', indicator = True)

df_merged_large
```

```python
#verificando ls frecuencia de los datos 
df_merged_large['_merge'].value_counts()
```

```python
# Verificando que se haya merged toda la data, como se ce cuando se hace con outer, quedan valores fuera del merge


df_merged_large = df_ords.merge(df_ords_prior, on = 'order_id', how = 'outer', indicator = True)

df_merged_large['_merge'].value_counts()
```

How many rows and columns do the dataframes you want to combine contain?

Do the dataframes contain information on the same subject? Do they share a common column or columns? 

Should your combined dataframe be long format or wide format? 

Do you expect a full match after combining the dataframes?
