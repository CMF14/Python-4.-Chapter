## 3.- Merging both Dataframe

### Suitable way to combine both dataframe 


Possible merge of both dataframes by using product_id as key variable

suitable merge form: left

After merge: verify result.



```python
df_ords_prods_merge = df_orders_products_combined.merge(df_prods, on='product_id')

df_ords_prods_merge.head()
```

Confirming the Results of the Merge Using a Merge Flag

```python

df_ords_prods_merge = df_orders_products_combined.merge(df_prods, on='product_id', indicator=True)

df_ords_prods_merge.head()
```

```python
# Changing the name of the merge column to avoid conflicts in the future!
df_ords_prods_merge = df_orders_products_combined.merge(df_prods, on='product_id', how='left', indicator='merge_check')
```
