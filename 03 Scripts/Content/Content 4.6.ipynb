## Concatenating Data

```python
#ahora vamos a concatenar un long dataframe


# Define a dictionary containing January 2020 data
data1 = {'customer_id':['6732', '767', '890', '635'],
    'month':['Jan-20', 'Jan-20', 'Jan-20', 'Jan-20'],
    'purchased_meat':[0, 13, 3, 4],
    'purchased_alcohol':[1, 2, 10, 0],
    'purchased_snacks':[10, 5, 1, 7]}

# Define a dictionary containing February 2020 data
data2 = {'customer_id':['6732', '767', '890', '635'],
    'month':['Feb-20', 'Feb-20', 'Feb-20', 'Feb-20'],
    'purchased_meat':[0, 10, 5, 3],
    'purchased_alcohol':[2, 4, 14, 0],
    'purchased_snacks':[15, 3, 2, 6]}
```

```python
#Convert the dictionary into dataframe

#df tiene datos de enero 2020
df = pd.DataFrame(data1,index=[0, 1, 2, 3])

#df_1 tiene fatos de febrero 2020

df_1 = pd.DataFrame(data2,index=[0, 1, 2, 3])
```

```python
df
```

```python
df_1
```

```python
## Ahora vamos a concatenar las dos df - por medio de:

# con frames se crea una lista que contiene los dos df
frames = [df, df_1]

# y luego se concatenan usando .concat
df_concat = pd.concat(frames)
```

```python
#ahora tenemos un marco de datos que contiene los dos meses, enero y febrero. o long dataframe 
df_concat
```

```python
#ten en cuenta que esta ultima opción es para hacer un long-dataframe, por lo que asumimos que el Axis=0 
# y no lo incluimos en la formula, porque es predeterminado. 
#pero si agregamos un axis=1, entonces python concatena los dos df uno al lado del otro. 
#ejemplo:_
```

```python
df_concat = pd.concat(frames, axis = 1)
#agregando en axis=1 en el código, obtenemos un frame ancho. 
```

## Merging Data

```python
#df.merge()

# tipos de uniones
#hay inner, solo cuando se unen las intercepciones de las dos datas 
# lef=  Una unión izquierda conserva la totalidad de un marco de datos y lo combina con la parte de intersección entre los dos. 

# right = Similar a una unión izquierda, una unión derecha conserva la totalidad de un marco de datos y lo combina con la parte de intersección entre los dos.
# full outer join= se usa principalmente cuando los marcos de datos son del mismo tamao, de lo contratio habrían problemas en la localización de los datos y que estos estén completos


#df.merge incluye también un argumentos importante: on, que es la columna identificadora comun entre los datos.
#normalmente el id del cliente o alguna referencia que se encuentre en las dos tablas.
```

```python
#Aquí un ejemplo de como aplicar df.merge
#primero creemos una data:
data3 = {'customer_id':['6732', '767', '890', '635'],
    'month':['Jan-20', 'Jan-20', 'Jan-20', 'Jan-20'],
    'days_purchased_on':[0, 10, 4, 1]}

#ahora hacemos un df:

df_2=pd.DataFrame(data3, index=[0,1,2,3] )

#Así se ve df_2:

df_2
```

```python
#Ahora let merge estos df_1 y df_2 usando el customer_id como argumento on:

df_merged = df.merge(df_2, on = 'customer_id')

#así se vería=
df.merged

#resultado una un wide frame
```

```python
#comentario= tenemos 2 months - month x y month y. así se vería la tabla si colocamos
#también la variable month como indicador:

df_merged = df.merge(df_2, on = ['customer_id', 'month'])

#así quedarái solo una columna month

df_merged
```

```python
# ahora que pasa si en una  de las listas los datos no están completamente compaginados?
# para verificar que todos los datos estén iguales, se coloca un incidador adicional de la siguiente forma:

#ejemplö: WICHTIG!!!

df_merged = df.merge(df_2, on = ['customer_id', 'month'], indicator=True)

df_merged

#en este caso todos los datos están incluidos en las dos listas y hay concidencia completa
```
